**CITY ANALYSIS**

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/content/Dataset .csv')
df.head()
```

```python
print(df.columns)
```

```python
city_restaurant_count = df.groupby('City').size().reset_index(name='Restaurant_Count')
```

```python
top_city = city_restaurant_count.loc[city_restaurant_count['Restaurant_Count'].idxmax()]
print(f"The city with the highest number of restaurants is {top_city['City']} with {top_city['Restaurant_Count']} restaurants.")
```

```python
city_avg_rating = df.groupby('City')['Aggregate rating'].mean().reset_index(name='Average_Rating')
print(city_avg_rating)
```

```python
city_avg_rating.sort_values(by='Average_Rating', ascending=False).plot(kind='bar', x='City', y='Average_Rating', figsize=(20, 15))
plt.title('Average Restaurant Rating per City')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.show()
```
