```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/content/Dataset .csv')
```

```python
print(df.columns)
```

```python
df['Price range'].unique()
```

```python
plt.figure(figsize=(7, 3))
plt.hist(df['Price range'], bins=10, edgecolor='k', alpha=0.7)
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python

price_range_counts = df['Price range'].value_counts()
price_range_percentage = df['Price range'].value_counts(normalize=True) * 100

price_range_distribution = pd.DataFrame({'Count': price_range_counts, 'Percentage': price_range_percentage})

price_range_distribution
```

```python
plt.figure(figsize=(7, 3))
price_range_distribution['Percentage'].plot(kind='bar', color='skyblue')
plt.title('Percentage of Restaurants by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
