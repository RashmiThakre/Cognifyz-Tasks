```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('/content/Dataset .csv')
```

```python
print(df.columns)
```

```python
online_delivery_counts = df['Has Online delivery'].value_counts()
online_delivery_percentage = df['Has Online delivery'].value_counts(normalize=True) * 100

online_delivery_distribution = pd.DataFrame({'Count': online_delivery_counts, 'Percentage': online_delivery_percentage})
online_delivery_distribution
```

```python
avg_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean().reset_index()
avg_ratings.columns = ['Has Online delivery', 'Aggregate rating']
```

```python
sns.barplot(x='Has Online delivery', y='Aggregate rating', data=avg_ratings)
plt.title('Average Ratings of Restaurants With and Without Online Delivery')
plt.xlabel('Has Online Delivery')
plt.ylabel('Average Rating')
plt.show()
```
